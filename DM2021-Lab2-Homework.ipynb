````python
from tensorflow import keras
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense
from tensorflow.keras.layers import ReLU, Softmax
from tensorflow.keras.callbacks import CSVLogger
from sklearn.metrics import accuracy_score

label_encoder = LabelEncoder()
label_encoder.fit(y_train)

def label_encode(le, labels):
    enc = le.transform(labels)
    return keras.utils.to_categorical(enc)

def label_decode(le, one_hot_label):
    dec = np.argmax(one_hot_label, axis=1)
    return le.inverse_transform(dec)

y_train_ = label_encode(label_encoder, y_train)
y_test_ = label_encode(label_encoder, y_test)

input_shape = X_train.shape[1]
output_shape = len(label_encoder.classes_)
````

````python
import emoji
import string
import re
from nltk.tokenize import TweetTokenizer
from nltk.stem.porter import PorterStemmer
from nltk.stem import WordNetLemmatizer
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_extraction.text import TfidfVectorizer
from tensorflow.keras.preprocessing.text import Tokenizer
from sklearn.feature_extraction.text import CountVectorizer
````

#### Remove Patterns

Hashtags, user tags (ex:@Alan, not sure what's the name), numbers and puncuations are removed using this function

````python
def remove_patterns(txt, patterns):
    for pattern in patterns:
        txt = re.sub(pattern,'', txt)
    return txt
````

#### Tokenize
 text (a sentence) ---tokenize--> words (list of words)

````python
def tokenize(text_df):
    tweet_tokenizer = TweetTokenizer(reduce_len = True)
    words = text_df.apply(lambda x :tweet_tokenizer.tokenize(x))
    return words
````

#### Lemmatize

I use lemmatizing instead of stemming since i initially attempted using a pretraind GloVe as embedding layer.  
A word stemmed won't be in the vocab list of Glove.

````python
def lemmatize(txt_list):
    lemmatizer = WordNetLemmatizer()
    return [lemmatizer.lemmatize(w) for w in txt_list]
````

#### Word Segmentation (Hashtags)

Hashtags are some combinations of words.Therefore i used word segementaion against the hashtags and append the segmmentation behind the tokenized text.

````python
import wordsegment
wordsegment.load()
def hashtag_segmentation(pd_series):
    segmented  = []
    for i,words in enumerate(pd_series.iloc,1):
        print(f'\r{i}/{len(pd_series)}',end = '')
        words_ = []
        for word in words:
            words_.extend(wordsegment.segment(word))
        segmented.append(words_)
    print('\nDone')
    return segmented
````

#### Remove Stop Words

````python
def remove_stopwords(words,stopwords = None):
    if stopwords is None:
        stopwords = stopwords = nltk.corpus.stopwords.words('english')
    return  [_ for _ in words if _ not in stopwords]    
````

#### Run All Preprocessings on Both train and test data

````python
def do_all(df):
    print('Removing patterns...')
    patterns = [r'#\w+',r'@\w+','<LH>','\d+', f"[{string.punctuation}]"]     
    df['text'] = df['text'].apply(lambda x: remove_patterns(x,patterns).lower())
    # nothing but just reducing the spaces between words in text
    df['text'].apply(lambda x :' '.join(x.split())) 
    # --tokenize-------------
    print('Tokenizing...')
    df['words'] = tokenize(df['text'])
    # --Hashtag Segmentation-
    print('Segmenting Hashtags...')
    df['segmented_hashtags'] = hashtag_segmentation(df.hashtags)
    for r in df.iloc:
        r.words.extend(r.segmented_hashtags)
    # --remove stop words----
    print('Removing stop words...')
    stopwords = nltk.corpus.stopwords.words('english')
    df['words'] = df['words'].apply(lambda x : remove_stopwords(x,stopwords = stopwords))
    # --Lemmatize------------
    print('Lemmatizing...')
    df['words'] = df['words'].apply(lambda x :lemmatize(x)) 
    print('Done!')
    return df
````

### Prediction Model

````python
from tensorflow.keras.layers import Input, Activation
from tensorflow.keras.layers import BatchNormalization,Dropout
from tensorflow.keras.layers import Dense,Conv1D,MaxPooling1D,LSTM
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import CSVLogger,ModelCheckpoint
from sklearn.model_selection import train_test_split
````

````python
def build_model(input_shape=(1000,), output_shape=8, lr = 0.001):
    input_layer = Input(shape=input_shape)
    x=Conv1D(filters=256, kernel_size=3,padding='same',activation='relu')(input_layer)   
    x = MaxPooling1D(pool_size=2)(x)
    x = LSTM(units=256,unroll = True, recurrent_initializer='orthogonal',
                 return_sequences=False,
                   dropout=0.2, recurrent_dropout=0.2)(x)
    x = BatchNormalization()(x)
    
    x = Dense(units=256)(x)
    x = BatchNormalization()(x)
    x  = Activation('relu')(x)
    x = Dropout(0.2)(x)

    output_layer = Dense(output_shape, activation = 'softmax')(x) 
    model = Model(input_layer, output_layer)
    model.compile(optimizer = Adam(lr),
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])
    model.summary()
    return model
````
